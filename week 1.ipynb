```python
import matplotlib.pyplot as plt

# -------------------------------------------------
# 1. New Sample 10-Day Dataset
# -------------------------------------------------
days = list(range(1, 11))

opening_price = [150, 152, 151, 153, 155, 154, 156, 158, 157, 159]
closing_price = [152, 150, 153, 154, 156, 152, 157, 159, 155, 160]

# -------------------------------------------------
# 2. Compute Daily Percentage Change
# -------------------------------------------------
percentage_changes = []

for op, cp in zip(opening_price, closing_price):
    pct = ((cp - op) / op) * 100
    percentage_changes.append(pct)

# -------------------------------------------------
# 3. Apply BUY / SELL / HOLD Rules
# -------------------------------------------------
signals = []

for i in range(len(percentage_changes)):
    today = percentage_changes[i]
    yesterday = percentage_changes[i-1] if i > 0 else 0

    # BUY rules
    if (today > 2 and yesterday > 0):  
        signals.append("BUY")
    elif today > 0 and yesterday > 0:  
        signals.append("BUY")

    # SELL rules
    elif today < -2:  
        signals.append("SELL")
    elif today < 0 and yesterday < 0:  
        signals.append("SELL")

    # HOLD rules
    else:
        signals.append("HOLD")

# -------------------------------------------------
# 4. Print Table
# -------------------------------------------------
print("Day | Open | Close | % Change | Signal")
print("--------------------------------------------")

for i in range(10):
    print(f"{i+1:>3} | {opening_price[i]:>4} | {closing_price[i]:>5} | "
          f"{percentage_changes[i]:>8.2f}% | {signals[i]}")

# -------------------------------------------------
# 5. BAR CHART - Percentage Change
# -------------------------------------------------
plt.figure()
plt.bar(days, percentage_changes)
plt.title("Daily Percentage Change")
plt.xlabel("Day")
plt.ylabel("% Change")
plt.show()

# -------------------------------------------------
# 6. HISTOGRAM - Distribution of % Change
# -------------------------------------------------
plt.figure()
plt.hist(percentage_changes)
plt.title("Histogram of Percentage Change")
plt.xlabel("% Change")
plt.ylabel("Frequency")
plt.show()

# -------------------------------------------------
# 7. SCATTER PLOT - Opening vs Closing Price
# -------------------------------------------------
plt.figure()
plt.scatter(opening_price, closing_price)
plt.title("Scatter Plot: Opening vs Closing Price")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()
```
